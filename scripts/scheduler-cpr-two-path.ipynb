```python
## instantaneous goodput

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

i = 0


file = open('../results-wns3/scheduler/scheduler-133'+'/scheduler'+str(i)+'-queue.txt', 'r')
finish_time = float(file.readlines()[-1].split('\t')[0])
file = open('../results-wns3/scheduler/scheduler-133'+'/scheduler'+str(i)+'-rx.txt', 'r')
lines = file.readlines()
goodput = []
c_time = []
for line in lines:
    temp = line.split('\t')
    # if (temp[0] == '1' and c_time == []):
    c_time.append(float(temp[1]))
    c_time.append(float(temp[2])/1024/1024)
    # if (temp[0] == '3' and len(c_time) == 2):
        # c_time[1] = c_time[1] + float(temp[5])
    c_time.append(temp[0])
    c_time.append(i)
    goodput.append(c_time)
    if (c_time[0] > finish_time):
        break
    c_time = []
dataTotal = pd.DataFrame (goodput, columns = ['Time', 'goodput', 'path', 'scheduler'])        



# rr = help(0)  
# minrtt = help(1)
# blest = help(2)  
# ecf = help(3)
# peek = help(4)
# mab = help(5)  

# print(dataTotal)

toDrop =dataTotal.loc[dataTotal["path"] == '3'].index.tolist()
one = dataTotal.drop(toDrop)
toDrop =dataTotal.loc[dataTotal["path"] == '1'].index.tolist()
two = dataTotal.drop(toDrop)
# print(one)
```

```python

plt.figure(figsize=(8, 10))
plt.grid(linestyle="--")  # 
ax = plt.gca()
# ax.spines['top'].set_visible(False)  # 
# ax.spines['right'].set_visible(False)  # 


plt.plot(one['Time'], one['goodput'], color="green", label="P0", linewidth=6)
plt.plot(two['Time'], two['goodput'], color="red", label="P1", linewidth=6)
# plt.plot(blest['Time'], blest['goodput'], color="blue", label="BLEST", linewidth=4)
# plt.plot(ecf['Time'], ecf['goodput'], color="c", label="ECF", linewidth=4)
# plt.plot(peek['Time'], peek['goodput'], color="orange", label="Peekaboo", linewidth=4)

# group_labels = ['1-1', '1-2', '1-3', '1-4', '1-5', '1-6', '1-7', '1-8', '1-9']  # x
plt.xticks(fontsize=20, fontweight='bold')  # 10
plt.yticks(fontsize=20, fontweight='bold')
# plt.title("Medium Dynamicity", fontsize=12, fontweight='bold')  # 12
plt.ylabel("Received Bytes (MB)", fontsize=24, fontweight='bold')
plt.xlabel("Time (s)", fontsize=24, fontweight='bold')
plt.xlim(0, 14)  # x
plt.ylim(0, 5)
# plt.ylim(2, 12)

# plt.legend()          #
plt.legend(loc=0, numpoints=1)
leg = plt.gca().get_legend()
ltext = leg.get_texts()
plt.setp(ltext, fontsize=20, fontweight='bold')  # 

# plt.savefig('./goodput_low.pdf', format='pdf')  # svg,emfword
# plt.savefig('./goodput_medium.pdf', format='pdf')  # svg,emfword
plt.savefig('./received_bytes_scheduler'+str(i)+'.pdf', format='pdf',bbox_inches ="tight",pad_inches = 0,transparent = False)  # svg,emfword
plt.savefig('./received_bytes_scheduler'+str(i)+'.png', format='png')  # svg,emfword
plt.show()
```
