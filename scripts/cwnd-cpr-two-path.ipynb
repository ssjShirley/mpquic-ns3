```python
## instantaneous goodput

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

i = 0
dir = 'cwnd-0-4'
file = open('../results-wns3/cwnd-unlimited/'+dir+'/scheduler'+str(i)+'-cwnd-change-0.txt', 'r')
lines = file.readlines()
goodput = []
c_time = []
for line in lines:
    temp = line.split('\t')
    c_time.append(float(temp[0]))
    c_time.append(int(temp[1])/1460)
    c_time.append(int(temp[2])/1460)
    goodput.append(c_time)
    c_time = []
dataTotal0 = pd.DataFrame (goodput, columns = ['Time', 'cwnd1', 'cwnd2'])        


file = open('../results-wns3/cwnd-unlimited/'+dir+'/scheduler'+str(i)+'-cwnd-change-1.txt', 'r')
lines = file.readlines()
goodput = []
c_time = []
for line in lines:
    temp = line.split('\t')
    c_time.append(float(temp[0]))
    c_time.append(int(temp[1])/1460)
    c_time.append(int(temp[2])/1460)
    goodput.append(c_time)
    c_time = []
dataTotal1 = pd.DataFrame (goodput, columns = ['Time', 'cwnd1', 'cwnd2'])  

print(dataTotal1)

# toDrop =dataTotal.loc[dataTotal["path"] == '3'].index.tolist()
# one = dataTotal.drop(toDrop)
# toDrop =dataTotal.loc[dataTotal["path"] == '1'].index.tolist()
# two = dataTotal.drop(toDrop)
# print(one)
```

```python

plt.figure(figsize=(8, 6))
plt.grid(linestyle="--")  # 
ax = plt.gca()
# ax.spines['top'].set_visible(False)  # 
# ax.spines['right'].set_visible(False)  # 


plt.plot(dataTotal0['Time'], dataTotal0['cwnd2'], color="green", label="P0", linewidth=4)
plt.plot(dataTotal1['Time'], dataTotal1['cwnd2'], color="red", label="P1", linewidth=4)
# plt.plot(blest['Time'], blest['goodput'], color="blue", label="BLEST", linewidth=4)
# plt.plot(ecf['Time'], ecf['goodput'], color="c", label="ECF", linewidth=4)
# plt.plot(peek['Time'], peek['goodput'], color="orange", label="Peekaboo", linewidth=4)

# group_labels = ['1-1', '1-2', '1-3', '1-4', '1-5', '1-6', '1-7', '1-8', '1-9']  # x
plt.xticks(fontsize=18, fontweight='bold')  # 10
plt.yticks(fontsize=18, fontweight='bold')
# plt.title("Medium Dynamicity", fontsize=12, fontweight='bold')  # 12
plt.ylabel("Congestion Window (Segment)", fontsize=20, fontweight='bold')
plt.xlabel("Time (s)", fontsize=20, fontweight='bold')
# plt.xlim(4, 15)  # x
# plt.ylim(4, 14)
# plt.ylim(2, 12)

# plt.legend()          #
plt.legend(loc=0, numpoints=1)
leg = plt.gca().get_legend()
ltext = leg.get_texts()
plt.setp(ltext, fontsize=20, fontweight='bold')  # 

# plt.savefig('./goodput_low.pdf', format='pdf')  # svg,emfword
# plt.savefig('./goodput_medium.pdf', format='pdf')  # svg,emfword
plt.savefig('./cwnd_newreno.pdf', format='pdf')  # svg,emfword
plt.savefig('./cwnd_newreno.png', format='png')  # svg,emfword
plt.show()
```
